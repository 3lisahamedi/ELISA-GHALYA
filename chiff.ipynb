```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Hypotheses
hypotheses = pd.DataFrame({
    "Hypothese": [
        "Petit volume (quelques milliers d'etudiants, quelques Go a dizaines de Go)",
        "Scoring hebdomadaire ou mensuel",
        "Petite equipe utilisatrice (scolarite / responsables pedagogiques)",
        "Usage prevention/accompagnement (pas sanction)"
    ]
})
print("Hypotheses")
display(hypotheses)

# Besoins techniques
besoins_tech = pd.DataFrame([
    {"Composant": "Stockage", "Description": "Donnees sources, final.csv, exports, resultats", "Niveau": "Faible"},
    {"Composant": "Calcul", "Description": "Execution pipeline (preprocess + scoring) planifie", "Niveau": "Moyen"},
    {"Composant": "Base de donnees", "Description": "Historique des scores (suivi dans le temps)", "Niveau": "Faible a moyen"},
    {"Composant": "Monitoring/logs", "Description": "Logs + alertes en cas d'echec", "Niveau": "Faible"},
    {"Composant": "Dashboard", "Description": "Notebook (gratuit) ou BI (variable)", "Niveau": "Variable"},
])
print("Besoins techniques")
display(besoins_tech)

# Estimation des couts mensuels (ordre de grandeur, sans Azure)
couts = pd.DataFrame([
    {"Poste": "Stockage", "Cout_min_eur": 0,  "Cout_max_eur": 5},
    {"Poste": "Calcul (serveur/VM legere)", "Cout_min_eur": 10, "Cout_max_eur": 40},
    {"Poste": "Base de donnees (petite)", "Cout_min_eur": 5,  "Cout_max_eur": 30},
    {"Poste": "Monitoring/logs", "Cout_min_eur": 0,  "Cout_max_eur": 10},
    {"Poste": "Dashboard (notebook)", "Cout_min_eur": 0,  "Cout_max_eur": 0},
])
couts["Cout_moyen_eur"] = ((couts["Cout_min_eur"] + couts["Cout_max_eur"]) / 2).round(2)

total_min = int(couts["Cout_min_eur"].sum())
total_max = int(couts["Cout_max_eur"].sum())
total_moy = float(couts["Cout_moyen_eur"].sum().round(2))

print("Estimation mensuelle (ordre de grandeur)")
display(couts)

print("Total mensuel minimum :", total_min, "EUR/mois")
print("Total mensuel maximum :", total_max, "EUR/mois")
print("Total mensuel moyen :", total_moy, "EUR/mois")

# Graphique repartition (cout moyen)
plt.figure(figsize=(8,4))
plt.bar(couts["Poste"], couts["Cout_moyen_eur"])
plt.title("Repartition des couts mensuels (moyenne, hors BI payant)")
plt.xlabel("Poste")
plt.ylabel("Cout moyen (EUR/mois)")
plt.xticks(rotation=25, ha="right")
plt.show()

# Besoins humains (jours-homme)
besoins_humains = pd.DataFrame([
    {"Role": "Data Engineer", "Charge_jours": "5-10", "Taches": "Pipeline automatise, base, securite"},
    {"Role": "Data Scientist", "Charge_jours": "7-12", "Taches": "Modeles, validation, metriques"},
    {"Role": "Data Analyst", "Charge_jours": "3-6", "Taches": "Dashboard KPI, restitution"},
    {"Role": "RGPD/DPO", "Charge_jours": "1-3", "Taches": "Conformite, registre, conservation"},
    {"Role": "Metier", "Charge_jours": "2-4", "Taches": "Regles d'action, validation"},
])
print("Besoins humains (mise en place initiale)")
display(besoins_humains)

print("Conclusion a copier :")
print("Le cout principal est la mise en place initiale. Ensuite, la maintenance est estimee a 1-2 jours par mois.")
```
